## Gradient Descent

```python
# we need some gradient: starting point, learning rate, temporary value
```

```python
#Gradient Descent
new_x=3 #starting point #random guess
previous_x=0  #temporary value for keeping loop
step_multiplier=0.1 #learning rate


for n in range(500):
    previous_x=new_x
    gradient=df(previous_x)
    new_x= previous_x-step_multiplier*gradient
    
print('Local minimum occurs at:',new_x)
print('slope or df(x) value at this point is: ' ,df(new_x))
print('f(x) value or cost at this point is:' ,f(new_x))
```

```python
# include the level of precision that you wanna met

new_x=3 #starting point #random guess
previous_x=0  #temporary value
step_multiplier=0.1 #learning rate
precision= 0.00001


x_list= [new_x] #list for x
slope_list=[df(new_x)] #list for slope




for n in range(500):
    previous_x=new_x
    gradient=df(previous_x)
    new_x= previous_x-step_multiplier*gradient
    
    step_size=abs(new_x-previous_x)
    #print(step_size)
    
    x_list.append(new_x) #append the value
    slope_list.append(df(new_x))
    
    
    if step_size < precision:
        print('Loop ran this many time:',n)
        break

    
print('Local minimum occurs at:',new_x)
print('slope or df(x) value at this point is: ' ,df(new_x))
print('f(x) value or cost at this point is:' ,f(new_x))
```

```python
#superimpose the gradient descent calculations on plot

plt.figure(figsize=[15,5]) #width and height

# 1chart: cost function
plt.subplot(1,2,1) #row,columns,index

plt.xlim([-3,3])
plt.ylim(0,8)

plt.title('cost function',fontsize=17)
plt.xlabel('X',fontsize=16)
plt.ylabel('f(x)',fontsize=16)

plt.plot(x_1,f(x_1),color='blue',linewidth=3)

values=np.array(x_list)
plt.scatter(x_list,f(values), color='red', s=100, alpha=0.6)


# 2chart: Derivatives
plt.subplot(1,2,2) 

plt.xlim(-2,3)
plt.ylim(-3,6)
plt.grid()

plt.title('slope of the cost function',fontsize=17)
plt.xlabel('X',fontsize=16)
plt.ylabel('f(x)',fontsize=16)


plt.plot(x_1,df(x_1),color='skyblue',linewidth=5)
plt.scatter(x_list,slope_list, color='red', s=100, alpha=0.5)


plt.show()
```

## Gradient Descent as a python function

```python
#gradient descent takes 4 arguments--> the derivative function,value for initial guess,learning rate,precision
#gradient-> value of the slope of the cost/ mulitiplier&gradient-> determine how big of a step we take


def gradient_descent(derivative_func,initial_guess,multiplier=0.02,precision=0.001,
                    max_iter=100): #default/optional value
    new_x=initial_guess                       #starting point #random guess
    x_list= [new_x] #list for x
    slope_list=[derivative_func(new_x)] #list for slope

    for n in range(max_iter):
        previous_x=new_x
        gradient=derivative_func(previous_x)
        new_x= previous_x-multiplier*gradient
    
        step_size=abs(new_x-previous_x)
        x_list.append(new_x) #append the value
        slope_list.append(derivative_func(new_x))
    
        if step_size < precision:
            break
    return new_x, x_list,slope_list
```

```python
local_min,list_x,deriv_list= gradient_descent(dg, 0.5, 0.02, 0.001)
print('Local min occurs at:',local_min)
print('Number of steps:',len(list_x))
```

# Example 4 -Data viz with 3D charts


## Minimize $$f(x,y)= \frac{1}{3^{-x^2-y^2}+1}$$


Minimize $$f(x,y) = \frac {1}{r+1}$$ where $r$ is $3^{-x^2-y^2}$


```python
#Create a function
```

```python
def f(x,y):
    r=3**(-x**2 - y**2)
    return 1/(r +1)
```

```python
#create a chart for this function
```

```python
#Make our x,y data
x_4=np.linspace(start=-2,stop=2,num=200)
y_4=np.linspace(start=-2,stop=2,num=200)

print('Shape of X array',x_4.shape)


# convert a x_4 from an array to a 2D metrix
# from numpy package use meshgrid -> Return coordinate matrices from coordinate vectors.

x_4, y_4 = np.meshgrid(x_4, y_4)
print('Array after meshgrid',x_4.shape)
```

```python
#Generating 3D plot
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
 #gca-->get current axes->getting the axes from the figure and storing in the object 
fig = plt.figure(figsize = [16,14])
#ax = fig.gca(projection='3d') -> get an error
ax = fig.add_subplot(projection='3d')

ax.set_xlabel('x',fontsize=20)
ax.set_ylabel('y',fontsize=20)
ax.set_zlabel('f(x,y) - cost',fontsize=20)


# Now we can plot our function on it.  #check the documentation ax.plot_surface we need to have a 2D but x_4 and y_4 is a vector
ax.plot_surface(x_4, y_4, f(x_4,y_4), cmap=cm.winter, alpha=0.3)
plt.show()
```
